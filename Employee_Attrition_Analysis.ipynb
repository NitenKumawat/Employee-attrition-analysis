# Employee Attrition Analysis

Employee attrition analysis is a type of behavioral analysis where we study the behavior and characteristics of employees who left the organization and compare their characteristics with the current employees to find the employees who may leave the organization soon.

```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv("/content/EmployeeAttrition.csv")
```

```python
df.head()
```

```python
df.isnull() # return true if find missing values else false.
```

```python
df.isnull().sum() # return total entries of missing values in dataframe.
```

# EDA (Exploratory Data analysis).

```python
import matplotlib.pyplot as plt # virtualization library.
import seaborn as sns # update virsion of matplotlib library.
```

```python
import plotly.graph_objects as go # plotly is 3d virtualization library.
import plotly.io as pio
pio.templates.default = "plotly_white"
```

filtter the data to show only "yes" values in the "Attrition" column.

```python
attr_df= df[df["Attrition"]=="Yes"] # using this method we extract those employees who leave the orgnization.
```

```python
attr_df
```

# Attrition by department.


```python
attr_dpt=attr_df.groupby(["Department"]).size().reset_index(name="count")
attr_dpt
```

```python
fig= go.Figure(data=[go.Pie(
    labels= attr_dpt["Department"],
    values= attr_dpt["count"],
    hole= 0.4,
    marker= dict(colors = ["green", "yellow"]),
    textposition= "inside"
)])
fig.update_layout(title="Attrition by Department")
fig.show()
```

# Attrition by EducationField

```python
attr_EF=attr_df.groupby(["EducationField"]).size().reset_index(name="count")
attr_EF
```

```python
fig= go.Figure(data=[go.Pie(
    labels= attr_EF["EducationField"],
    values= attr_EF["count"],
    hole= 0.4,
    marker= dict(colors = ["green", "yellow"]),
    textposition= "inside"
)])
fig.update_layout(title="Attrition by EducationField")
fig.show()
```

# Percentage of attrition by number of years at the company.

```python
attr_years_total=attr_df.groupby(["YearsAtCompany"]).size().reset_index(name="count")
attr_years_total
```

# During promotion how many employees left the orgainzation.

```python
attr_promo=attr_df.groupby(["YearsSinceLastPromotion"]).size().reset_index(name="count")
attr_promo
```

# Percebtage of attrition by gender

```python
attr_gender=attr_df.groupby(["Gender"]).size().reset_index(name="count")
attr_gender
```

```python
fig= go.Figure(data=[go.Pie(
    labels= attr_gender["Gender"],
    values= attr_gender["count"],
    hole= 0.4,
    marker= dict(colors = ["green", "yellow"]),
    textposition= "inside"
)])
fig.update_layout(title="Attritio by gender")
fig.show()
```
